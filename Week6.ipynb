### How to:
* create a text file listing the accessions of all of the genomes
* upload text file into [Batch Entrez](https://www.ncbi.nlm.nih.gov/sites/batchentrez), search for "Assembly"
* download files from GenBank as Genomic FASTA files
* unzip files from .fna.gz files to .fna files (click to open .fna.gz)
* run prodigal to convert .fna files to .faa files (nucleotide sequence to amino acid sequence)
* use kmerCounter.R to convert .faa files into kmer counts (amino acid sequence to kmer counts)
* use multigenomeCSVGenerator.py to choose specific kmer counts and combine into one large matrix (kmer counts vectors to selected kmer count matrix)
* use canopyClustering.py to group proteins (kmer count matrix into groups of proteins)
* use grouping.R to visualize groups and get core/accessory/singleton composition for pangenome
