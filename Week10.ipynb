Today, I also worked to fix a bug that I had found for the pan-genome evolution plot. The plot is generated by sampling a few genomes at a time and getting the average pan-genome composition for each step in the graph. The sampling was being performed with replacement though, and this was skewing the singleton numbers on the graphs. I changed the sampling algorithm to sample without replacement, and this seemed to fix the graphs.

## Tuesday July 23
Today, I looked at trying to use Prokka to annotate the genes from Prodigal. Prokka actually uses Prodigal to discover genes from input .fna genome files, but it uses specific settings. When I ran prodigal for gene finding, I simply called `prodigal -i ___.fna -a ____.faa -q`. But when Prokka uses Prodigal, it uses `prodigal -i ___.fna -a ___.faa -q -c -m -g 11 -p single -f sco`. These settings make Prodigal use standard bacterial/archael reference genomes (-g 11), prohibit partial genes at ends of sequences (-c), and output data using simple coordinate output (-f sco). I wondered what effect the partial genes at ends of sequences might have on pan-genome construction, so I decided to run Prodigal with the new settings to remove partial genes, and then I constructed pan-genomes from the results. For both *Staphylococcus aureus* and *E. coli/Shigella spp.*, the number of singleton genes decreased, while the rest of the pan-genome stayed in roughly the same proportions. Results:
